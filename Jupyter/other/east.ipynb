```python
from bs4 import BeautifulSoup
import time
import requests
import csv

page = 1
A=[]
url = 'http://www.backpackers.com.tw/forum/'

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}
pageurl = 'http://www.backpackers.com.tw/forum/forumdisplay.php?f=55&prefixid=voyage&sort=dateline&order=desc&daysprune=-1'
res = requests.get(pageurl, headers=head)
soup = BeautifulSoup(res.text.encode("utf-8"),'lxml') 

while page<=72:
    i=1
    print('')
    print('-----page=',page)
    for elem in soup.select('#threadbits_forum_55 tr')[1:]:
        #print(i,elem.select('.alt1 div a')[-1]['href'])
        print(i,end=' ')
        if len(elem.select('.deleted div a'))<1:
            go = elem.select('.alt1 div a')[-1]['href']
        else:
            go = elem.select('.deleted div a')[-1]['href']
        goto = url+go
        A.append([str(page)+'-'+str(i),goto])
        i+=1
        time.sleep(0.1)
    time.sleep(3)
    page+=1
    nxtpage = pageurl+'&page='+str(page)
    res = requests.get(nxtpage, headers=head)
    soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')

with open('關西url.csv','w+',newline='') as f:
    writer = csv.writer(f)
    writer.writerows(A)
print('finished!!!')      
```

```python
import csv
from bs4 import BeautifulSoup
import requests
import time

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}
session = requests.Session()
fin=[]
fin.clear()
url=[]

with open('關西url.csv','r') as f:
    for i in csv.reader(f):
        url.append(i[1])

for t in range(len(url)):
    print(t+1,end=' ')
    if (t+1)%30==0:
        print('')
    res = session.get(url[t], headers=head)
    soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')
    title = soup.select('#posts .alt1 .smallfont')[0].contents[3].text.strip()
    time.sleep(0.25)
    date = soup.select('#posts .alt1 .smallfont')[0].contents[8].strip()
    time.sleep(0.25)
    content = soup.select('#posts .alt1 .vb_postbit')[0].text.strip()
    time.sleep(0.25)
    fin.append([title,date,content,url[t]])
    time.sleep(0.25)

with open('關西5.csv','w',encoding='utf-8',newline='') as f:
    writer = csv.writer(f)
    writer.writerows(fin)
print('finished!!!')
            
#print(soup.select('#posts .alt1 .smallfont')[0].contents[3].text.strip()) #標題
#print(soup.select('#posts .alt1 .smallfont')[0].contents[8].strip()) #日期
#print(soup.select('#posts .alt1 .vb_postbit')[0].text.strip()) #內文
```
