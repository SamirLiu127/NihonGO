# 1. 抓網址

```python
#抓取PTT Japan_Travel 標題為"遊記"的網址
import requests
from bs4 import BeautifulSoup
import re
import csv

pat='遊記'
url=[]

#PTT 從4602到90頁 每次-1
for i in range(4602,89,-1):
    res=requests.get('https://www.ptt.cc/bbs/Japan_Travel/index'+str(i)+'.html')
    soup = BeautifulSoup(res.text.encode("utf-8"),'html.parser')
    print(i,end=' ')
    for line in soup.select('.r-ent'):
#         print(line.select('.title')[0].text)
        title = line.select('.title')[0].text
#         print(line.select('.date')[0].text)
        if bool(re.search(pat,title)):
#             print(line.select('.title a')[0]['href'])
            a = line.select('.title a')[0]['href']
            url.append(a)

with open('ptturl.csv','w+',newline='') as f:
    writer = csv.writer(f)
    writer.writerows(url)
print('finished!!!') 
```

```python
#TEST單篇文章(含日期轉換)
import csv
from bs4 import BeautifulSoup
import requests
import re

fin=[]
fin.clear()
url = 'https://www.ptt.cc/bbs/Japan_Travel/M.1485219082.A.B41.html'
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

res = requests.get(url)
soup =  BeautifulSoup(res.text.encode('utf-8'),"html.parser")

main = soup.find(id='main-content')
metas = main.select('div.article-metaline')

#Title
title = metas[1].select('span.article-meta-value')[0].string

#Date
date = metas[2].select('span.article-meta-value')[0].string
 #月份轉換
for i,month in enumerate(months): #enumerate加入index
    date = date.replace(month,str(i+1))
date = date.split(' ') #空白切割
date = date[4]+'/'+date[1]+'/'+date[2]

#Content
content=''
for i in main:
    #開頭為 <div 或 <span捨棄
    if bool(re.search('<div',str(i))) or bool(re.search('<span',str(i))):
        continue
    else:
        content += str(i)
        
fin.append([title,date,content,url])
    
with open('ptt.csv','w+',newline='',encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(fin)
print('finished!!!')
```

# 2. 抓文章(匯出CSV)

```python
#全文章抓取
import csv
from bs4 import BeautifulSoup
import requests
import re

ptt='https://www.ptt.cc'
fin=[]
fin.clear()
go=[]
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
nourl=[]


with open('ptturl.csv','r') as f:
    for i in csv.reader(f):
        go.append(i[0])
        
for t in range(len(go)):
    print(t+1,end=' ')
    url = ptt+go[t]
    res = requests.get(url)
    soup = BeautifulSoup(res.text.encode("utf-8"),"html.parser")
    
    main = soup.find(id='main-content')
    metas = main.select('div.article-metaline')
    if len(metas)<1:
        nourl.append(url)
        continue
    
    #Title
    title = metas[1].select('span.article-meta-value')[0].string
    
    #Date
    date = metas[2].select('span.article-meta-value')[0].string
     #月份轉換
    for i,month in enumerate(months): #enumerate加入index
        date = date.replace(month,str(i+1)) #字串取代
    date = date.split(' ') #空白切割(回傳list)
    date = date[4]+'/'+date[1]+'/'+date[2]
    
    #Content
    content = ''
    for i in main:
        if bool(re.search('<div',str(i))) or bool(re.search('<span',str(i))):
            continue
        else:
            content += str(i)
        
    fin.append([title,date,content,url])
    
with open('ptt.csv','w+',newline='',encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerows(fin)
print('finished!!!')
```
