```python
#北海道56 關東43 九州138 東北139 沖繩140 四國196 抓各文章網址匯出CSV
from bs4 import BeautifulSoup
import time
import requests
import csv

page = 1
A = []
url = 'http://www.backpackers.com.tw/forum/'

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}
pageurl = 'http://www.backpackers.com.tw/forum/forumdisplay.php?f=43&prefixid=voyage&sort=dateline&order=desc&daysprune=-1'
res = requests.get(pageurl, headers=head)
soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')

while page<47:
    i=1
    print('')
    print('-----page=',page)
    for elem in soup.select('#threadbits_forum_43 tr')[1:]:
        if page==24 and i==3:
            i+=1
            continue
        elif page==40 and i==13:
            i+=1
            continue
        elif page==44 and i==19:
            i+=1
            continue
        print(i,end=' ')
        #print(elem.select('div a')[1]['href'])
        if len(elem.select('.alt1 div a'))>1:
            go = elem.select('.alt1 div a')[1]['href']
        else:
            if len(elem.select('.atl1 div a'))==1:
                go = elem.select('.alt1 div a')[0]['href']
            else:
                go = elem.select('.deleted div a')[0]['href']
        #go = elem.select('.alt1 div a')[1]['href']
        A.append([str(page)+'-'+str(i),url+go])
        i+=1
        time.sleep(0.5)
    page+=1
    nxturl = pageurl+'&page='+str(page)
    res = requests.get(nxturl, headers=head)
    soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')

with open('關東url.csv','w+',newline='') as f:
    writer = csv.writer(f)
    writer.writerows(A)
print('finished!!!')
```

```python
#載入 CSV 爬內文(一次約800筆文章)
import csv
from bs4 import BeautifulSoup
import requests
import time

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}
session = requests.Session()
fin=[]
url=[]

#匯入網址
with open('關東url.csv','r') as f:
    for i in csv.reader(f):
        url.append(i[1])

for t in range(len(url)):
    print(t+1,end=' ')
    if (t+1)%30==0:
        print('')
    elif (t+1)%500==0:
        time.sleep(600)
    res = session.get(url[t], headers=head)
    soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')
    title = soup.select('#posts .alt1 .smallfont')[0].contents[3].text.strip() #標題
    time.sleep(0.25)
    date = soup.select('#posts .alt1 .smallfont')[0].contents[8].strip() #發文日期
    time.sleep(0.25)
    content = soup.select('#posts .alt1 .vb_postbit')[0].text.strip() #內文
    time.sleep(0.25)
    fin.append([title,date,content,url[t]]) #網址
    time.sleep(0.25)

with open('關東.csv','w',encoding='utf-8',newline='') as f:
    writer = csv.writer(f)
    writer.writerows(fin)
print('finished!!!')
            
#print(soup.select('#posts .alt1 .smallfont')[0].contents[3].text.strip()) #標題
#print(soup.select('#posts .alt1 .smallfont')[0].contents[8].strip()) #日期
#print(soup.select('#posts .alt1 .vb_postbit')[0].text.strip()) #內文
```
