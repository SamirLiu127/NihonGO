```python
#北海道 抓各文章網址
from bs4 import BeautifulSoup
import time
import requests
import csv

page = 1
A = []
url = 'http://www.backpackers.com.tw/forum/'

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}
pageurl = 'http://www.backpackers.com.tw/forum/forumdisplay.php?f=56&prefixid=voyage&sort=dateline&order=desc&daysprune=-1'
res = requests.get(pageurl, headers=head)
soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')

while page<25:
    i=1
    print('')
    print('-----page=',page)
    for elem in soup.select('#threadbits_forum_56 tr')[1:]:
        if page==17 and i==1:
            i+=1
            continue
        elif page==18 and i==11:
            i+=1
            continue
        print(i,end=' ')
        #print(elem.select('div a')[1]['href'])
        if len(elem.select('.alt1 div a'))>1:
            go = elem.select('.alt1 div a')[1]['href']
        else:
            go = elem.select('.alt1 div a')[0]['href']
        #go = elem.select('.alt1 div a')[1]['href']
        A.append([str(page)+'-'+str(i),url+go])
        i+=1
        time.sleep(1)
    page+=1
    nxturl = pageurl+'&page='+str(page)
    res = requests.get(nxturl, headers=head)
    soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')

with open('北海道url.csv','w+',newline='') as f:
    writer = csv.writer(f)
    writer.writerows(A)
        
```

```python
import csv
from bs4 import BeautifulSoup
import requests
import time

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'}
session = requests.Session()
fin=[]
j=1

with open('北海道url.csv','r') as f:
    for i in csv.reader(f):
        url=i[1]
        print(i[0],end=' ')
        if j%10==0:
            print('')
        res = session.get(url, headers=head)
        soup = BeautifulSoup(res.text.encode("utf-8"),'lxml')
        
        title = soup.select('#posts .alt1 .smallfont')[0].contents[3].text.strip()
        date = soup.select('#posts .alt1 .smallfont')[0].contents[8].strip()
        content = soup.select('#posts .alt1 .vb_postbit')[0].text.strip()
        fin.append([title,date,content,url])
        time.sleep(2)
        j+=1

with open('北海道.csv','w',encoding='utf-8',newline='') as f:
    writer = csv.writer(f)
    writer.writerows(fin)
            
#print(soup.select('#posts .alt1 .smallfont')[0].contents[3].text.strip()) #標題
#print(soup.select('#posts .alt1 .smallfont')[0].contents[8].strip()) #日期
#print(soup.select('#posts .alt1 .vb_postbit')[0].text.strip()) #內文
```
